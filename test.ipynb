**Ergebnis ausgeben**
das Ergebnis von `model.predict(feature_df)` ist eine Array, um ein eindeutiges Ergebnis zu liefern wird in der folgenden<br>
Funktion das Label ausgegeben, welches am häufigsten vorgekommen ist

```python
def most_common_string(arr):
    frequencies = {}
    for element in arr:
        if element in frequencies:
            frequencies[element] += 1
        else:
            frequencies[element] = 1
    return max(frequencies, key=frequencies.get)

predictions = model.predict(feature_df)
most_common_prediction = most_common_string(predictions)
print(f"Prediction: {most_common_prediction}")
```
